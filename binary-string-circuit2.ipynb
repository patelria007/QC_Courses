```python
# All of this manipulates circuit with 2 qubits

# Imports needed libraries/tools
import qiskit as qk
from qiskit import ClassicalRegister, QuantumRegister, QuantumCircuit
from qiskit import execute, Aer
from qiskit import IBMQ
from qiskit.tools.visualization import plot_histogram

# Used for mpl drawing at the end
style = {'backgroundcolor':'#FB7685'}

# Creates quantum circuits
q = QuantumRegister(2)
c = ClassicalRegister(2)
qc = QuantumCircuit(q, c)
```

```python
# Measuring for |00> state
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])
```

```python
# Simulates the circuit & plots histogram
backend = Aer.get_backend('qasm_simulator')
job_sim = execute(qc, backend, shots=1024)
sim_result = job_sim.result()
measurement_result = sim_result.get_counts(qc)
print(measurement_result)
plot_histogram(measurement_result, color = 'lightgreen')
```

```python
qc.barrier(q)

# Flipping qubit to get |01> state
qc.x(q[1])
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])
```

```python
# Simulates the circuit & plots histogram
backend = Aer.get_backend('qasm_simulator')
job_sim = execute(qc, backend, shots=1024)
sim_result = job_sim.result()
measurement_result = sim_result.get_counts(qc)
print(measurement_result)
plot_histogram(measurement_result)
```

```python
qc.barrier(q)

# Resets qubits to |00> state
qc.reset(q)

# Flips qubit to get |10>
qc.x(q[0])
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])
```

```python
# Simulates the circuit & plots histogram
backend = Aer.get_backend('qasm_simulator')
job_sim = execute(qc, backend, shots=1024)
sim_result = job_sim.result()
measurement_result = sim_result.get_counts(qc)
print(measurement_result)
plot_histogram(measurement_result)
```

```python
qc.barrier(q)

# Flips qubit to get |11> state
qc.x(q[1])
qc.measure(q[0], c[0])
qc.measure(q[1], c[1])
```

```python
# Simulates the circuit & plots histogram
backend = Aer.get_backend('qasm_simulator')
job_sim = execute(qc, backend, shots=1024)
sim_result = job_sim.result()
measurement_result = sim_result.get_counts(qc)
print(measurement_result)
plot_histogram(measurement_result, color = 'lightgreen')
```

```python
# Circuit visual
qc.draw(output='mpl', style=style)
```
